# LEVEL 1 — BASELINE MODEL
(RESNET18, TRANSFER LEARNING)

PROJECT SETUP & FOLDERS

```python
import os

BASE_DIR = "/content/CIFAR10_Project"

folders = [
    "level_1_code",
    "level_2_code",
    "level_3_code",
    "level_4_code",
    "level_5_code",
    "models",
    "results"
]

os.makedirs(BASE_DIR, exist_ok=True)
for f in folders:
    os.makedirs(os.path.join(BASE_DIR, f), exist_ok=True)

print("Project structure created:")
print(os.listdir(BASE_DIR))
```

IMPORTS

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Subset
import matplotlib.pyplot as plt
import numpy as np
import random
```

CONFIGURATION & GPU CHECK

```python
BATCH_SIZE = 128
EPOCHS = 20
LR = 1e-3
SEED = 42
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

print("Device in use:", DEVICE)
```

DATA TRANSFORMS (LEVEL-1)

```python
train_transform_lvl1 = transforms.Compose([
    transforms.Resize(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(224, padding=4),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
])

test_transform_lvl1 = transforms.Compose([
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
])
```

DATASET LOADING & SPLIT (80-10-10)

```python
train_full_lvl1 = datasets.CIFAR10(
    root="./data", train=True, download=True,
    transform=train_transform_lvl1
)

test_dataset_lvl1 = datasets.CIFAR10(
    root="./data", train=False, download=True,
    transform=test_transform_lvl1
)

indices = list(range(len(train_full_lvl1)))
random.shuffle(indices)

train_idx_lvl1 = indices[:40000]
val_idx_lvl1 = indices[40000:]

train_dataset_lvl1 = Subset(train_full_lvl1, train_idx_lvl1)
val_dataset_lvl1 = Subset(train_full_lvl1, val_idx_lvl1)

train_loader_lvl1 = DataLoader(train_dataset_lvl1, batch_size=BATCH_SIZE, shuffle=True)
val_loader_lvl1 = DataLoader(val_dataset_lvl1, batch_size=BATCH_SIZE, shuffle=False)
test_loader_lvl1 = DataLoader(test_dataset_lvl1, batch_size=BATCH_SIZE, shuffle=False)

print("Train:", len(train_dataset_lvl1),
      "Val:", len(val_dataset_lvl1),
      "Test:", len(test_dataset_lvl1))
```

MODEL DEFINITION (RESNET18)

```python
model_lvl1 = models.resnet18(pretrained=True)

for param in model_lvl1.parameters():
    param.requires_grad = False

model_lvl1.fc = nn.Linear(model_lvl1.fc.in_features, 10)
model_lvl1 = model_lvl1.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_lvl1.fc.parameters(), lr=LR)

print("Level-1 ResNet18 initialized")
```

TRAINING LOOP (ACCURACY + LOSS)

```python
train_accs_lvl1, val_accs_lvl1 = [], []
train_losses_lvl1, val_losses_lvl1 = [], []

for epoch in range(EPOCHS):
    model_lvl1.train()
    correct, total, running_loss = 0, 0, 0

    for x, y in train_loader_lvl1:
        x, y = x.to(DEVICE), y.to(DEVICE)

        optimizer.zero_grad()
        out = model_lvl1(x)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)

    train_acc = correct / total
    train_loss = running_loss / len(train_loader_lvl1)

    train_accs_lvl1.append(train_acc)
    train_losses_lvl1.append(train_loss)

    model_lvl1.eval()
    correct, total, running_loss = 0, 0, 0

    with torch.no_grad():
        for x, y in val_loader_lvl1:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model_lvl1(x)
            loss = criterion(out, y)

            running_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            total += y.size(0)

    val_acc = correct / total
    val_loss = running_loss / len(val_loader_lvl1)

    val_accs_lvl1.append(val_acc)
    val_losses_lvl1.append(val_loss)

    print(f"Level-1 Epoch [{epoch+1}/{EPOCHS}] | "
          f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | "
          f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
```

TEST EVALUATION

```python
model_lvl1.eval()
correct, total = 0, 0

with torch.no_grad():
    for x, y in test_loader_lvl1:
        x, y = x.to(DEVICE), y.to(DEVICE)
        out = model_lvl1(x)
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)

test_accuracy_lvl1 = correct / total
print("LEVEL 1 FINAL TEST ACCURACY:", round(test_accuracy_lvl1, 4))
```

SAVE MODEL

```python
torch.save(
    model_lvl1.state_dict(),
    f"{BASE_DIR}/models/level_1_resnet18.pth"
)

print("Level-1 model saved")
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(train_accs_lvl1, label="Train Accuracy")
plt.plot(val_accs_lvl1, label="Validation Accuracy")
plt.title("Level 1 Accuracy Curve")
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_losses_lvl1, label="Train Loss")
plt.plot(val_losses_lvl1, label="Validation Loss")
plt.title("Level 1 Loss Curve")
plt.legend()

plt.savefig(f"{BASE_DIR}/results/level_1_training_curves.png")
plt.show()

print("Level-1 training curves saved")
```

# LEVEL 2 — INTERMEDIATE TECHNIQUES

(Data Augmentation + Regularization + Fine-Tuning + Ablation)

LEVEL-2 CONFIGURATION

```python
EPOCHS_LVL2 = 25
LR_LVL2 = 1e-4
WEIGHT_DECAY_LVL2 = 1e-4

print("Level-2 configuration loaded")
```

LEVEL-2 DATA AUGMENTATION PIPELINE

```python
train_transform_lvl2 = transforms.Compose([
    transforms.Resize(224),
    transforms.RandomCrop(224, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225]),
    transforms.RandomErasing(p=0.5)
])

test_transform_lvl2 = test_transform_lvl1  # reuse clean test transform

print("Level-2 augmentation defined")
```

LEVEL-2 DATASET & LOADERS

```python
train_full_lvl2 = datasets.CIFAR10(
    root="./data", train=True, download=False,
    transform=train_transform_lvl2
)

test_dataset_lvl2 = datasets.CIFAR10(
    root="./data", train=False, download=False,
    transform=test_transform_lvl2
)

train_dataset_lvl2 = Subset(train_full_lvl2, train_idx_lvl1)
val_dataset_lvl2   = Subset(train_full_lvl2, val_idx_lvl1)

train_loader_lvl2 = DataLoader(train_dataset_lvl2, batch_size=BATCH_SIZE, shuffle=True)
val_loader_lvl2   = DataLoader(val_dataset_lvl2, batch_size=BATCH_SIZE, shuffle=False)
test_loader_lvl2  = DataLoader(test_dataset_lvl2, batch_size=BATCH_SIZE, shuffle=False)

print("Level-2 dataloaders ready")
```

LEVEL-2 MODEL (PARTIAL FINE-TUNING)

```python
model_lvl2 = models.resnet18(pretrained=True)

# Freeze all layers
for param in model_lvl2.parameters():
    param.requires_grad = False

# Unfreeze last residual block
for param in model_lvl2.layer4.parameters():
    param.requires_grad = True

model_lvl2.fc = nn.Linear(model_lvl2.fc.in_features, 10)
model_lvl2 = model_lvl2.to(DEVICE)

criterion_lvl2 = nn.CrossEntropyLoss()

optimizer_lvl2 = optim.Adam(
    filter(lambda p: p.requires_grad, model_lvl2.parameters()),
    lr=LR_LVL2,
    weight_decay=WEIGHT_DECAY_LVL2
)

scheduler_lvl2 = optim.lr_scheduler.StepLR(
    optimizer_lvl2, step_size=7, gamma=0.1
)

print("Level-2 ResNet18 initialized with fine-tuning")
```

LEVEL-2 TRAINING LOOP (ACCURACY + LOSS)

```python
train_accs_lvl2, val_accs_lvl2 = [], []
train_losses_lvl2, val_losses_lvl2 = [], []

for epoch in range(EPOCHS_LVL2):
    model_lvl2.train()
    correct, total, running_loss = 0, 0, 0

    for x, y in train_loader_lvl2:
        x, y = x.to(DEVICE), y.to(DEVICE)

        optimizer_lvl2.zero_grad()
        out = model_lvl2(x)
        loss = criterion_lvl2(out, y)
        loss.backward()
        optimizer_lvl2.step()

        running_loss += loss.item()
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)

    train_acc = correct / total
    train_loss = running_loss / len(train_loader_lvl2)

    train_accs_lvl2.append(train_acc)
    train_losses_lvl2.append(train_loss)

    model_lvl2.eval()
    correct, total, running_loss = 0, 0, 0

    with torch.no_grad():
        for x, y in val_loader_lvl2:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model_lvl2(x)
            loss = criterion_lvl2(out, y)

            running_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            total += y.size(0)

    val_acc = correct / total
    val_loss = running_loss / len(val_loader_lvl2)

    val_accs_lvl2.append(val_acc)
    val_losses_lvl2.append(val_loss)

    scheduler_lvl2.step()

    print(f"Level-2 Epoch [{epoch+1}/{EPOCHS_LVL2}] | "
          f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | "
          f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
```

LEVEL-2 TEST EVALUATION

```python
model_lvl2.eval()
correct, total = 0, 0

with torch.no_grad():
    for x, y in test_loader_lvl2:
        x, y = x.to(DEVICE), y.to(DEVICE)
        out = model_lvl2(x)
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)

test_accuracy_lvl2 = correct / total
print("LEVEL 2 FINAL TEST ACCURACY:", round(test_accuracy_lvl2, 4))
```

# LEVEL 3 — ADVANCED ARCHITECTURE DESIGN

Custom CNN + SE Attention

LEVEL-3 CONFIGURATION

```python
EPOCHS_LVL3 = 30
LR_LVL3 = 1e-3

print("Level-3 configuration loaded")
```

LEVEL-3 DATA TRANSFORMS

```python
train_transform_lvl3 = transforms.Compose([
    transforms.Resize(64),
    transforms.RandomCrop(64, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.2, 0.2, 0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])

test_transform_lvl3 = transforms.Compose([
    transforms.Resize(64),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])

print("Level-3 transforms defined")
```

LEVEL-3 DATASET & LOADERS

```python
train_full_lvl3 = datasets.CIFAR10(
    root="./data", train=True, download=False,
    transform=train_transform_lvl3
)

test_dataset_lvl3 = datasets.CIFAR10(
    root="./data", train=False, download=False,
    transform=test_transform_lvl3
)

train_dataset_lvl3 = Subset(train_full_lvl3, train_idx_lvl1)
val_dataset_lvl3   = Subset(train_full_lvl3, val_idx_lvl1)

train_loader_lvl3 = DataLoader(train_dataset_lvl3, batch_size=BATCH_SIZE, shuffle=True)
val_loader_lvl3   = DataLoader(val_dataset_lvl3, batch_size=BATCH_SIZE, shuffle=False)
test_loader_lvl3  = DataLoader(test_dataset_lvl3, batch_size=BATCH_SIZE, shuffle=False)

class_names = train_full_lvl3.classes

print("Level-3 dataloaders ready")
```

SE ATTENTION BLOCK

```python
class SEBlock(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction),
            nn.ReLU(),
            nn.Linear(channels // reduction, channels),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y
```

CUSTOM CNN WITH SE ATTENTION

```python
class CustomCNN_SE(nn.Module):
    def __init__(self):
        super().__init__()

        self.block1 = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            SEBlock(64),
            nn.MaxPool2d(2)
        )

        self.block2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            SEBlock(128),
            nn.MaxPool2d(2)
        )

        self.block3 = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            SEBlock(256),
            nn.MaxPool2d(2)
        )

        self.classifier = nn.Sequential(
            nn.Linear(256 * 8 * 8, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, 10)
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)
```

LEVEL-3 MODEL INITIALIZATION

```python
model_lvl3 = CustomCNN_SE().to(DEVICE)

criterion_lvl3 = nn.CrossEntropyLoss()
optimizer_lvl3 = optim.Adam(model_lvl3.parameters(), lr=LR_LVL3)

print("Level-3 Custom CNN with SE Attention initialized")
```

LEVEL-3 TRAINING LOOP (ACCURACY + LOSS)

```python
train_accs_lvl3, val_accs_lvl3 = [], []
train_losses_lvl3, val_losses_lvl3 = [], []

for epoch in range(EPOCHS_LVL3):
    model_lvl3.train()
    correct, total, running_loss = 0, 0, 0

    for x, y in train_loader_lvl3:
        x, y = x.to(DEVICE), y.to(DEVICE)

        optimizer_lvl3.zero_grad()
        out = model_lvl3(x)
        loss = criterion_lvl3(out, y)
        loss.backward()
        optimizer_lvl3.step()

        running_loss += loss.item()
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)

    train_acc = correct / total
    train_loss = running_loss / len(train_loader_lvl3)

    train_accs_lvl3.append(train_acc)
    train_losses_lvl3.append(train_loss)

    model_lvl3.eval()
    correct, total, running_loss = 0, 0, 0

    with torch.no_grad():
        for x, y in val_loader_lvl3:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model_lvl3(x)
            loss = criterion_lvl3(out, y)

            running_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            total += y.size(0)

    val_acc = correct / total
    val_loss = running_loss / len(val_loader_lvl3)

    val_accs_lvl3.append(val_acc)
    val_losses_lvl3.append(val_loss)

    print(f"Level-3 Epoch [{epoch+1}/{EPOCHS_LVL3}] | "
          f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | "
          f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
```

LEVEL-3 TEST EVALUATION

```python
model_lvl3.eval()
correct, total = 0, 0
y_true_lvl3, y_pred_lvl3 = [], []

with torch.no_grad():
    for x, y in test_loader_lvl3:
        x, y = x.to(DEVICE), y.to(DEVICE)
        out = model_lvl3(x)
        preds = out.argmax(1)

        correct += (preds == y).sum().item()
        total += y.size(0)

        y_true_lvl3.extend(y.cpu().numpy())
        y_pred_lvl3.extend(preds.cpu().numpy())

test_accuracy_lvl3 = correct / total
print("LEVEL 3 FINAL TEST ACCURACY:", round(test_accuracy_lvl3, 4))
```

SOFT-VOTING ENSEMBLE FUNCTION

```python
import torch.nn.functional as F

def ensemble_predict_lvl4(
    model_lvl2,
    model_lvl3,
    model_lvl4,
    test_loader_224,
    test_loader_64
):
    y_true, y_pred = [], []

    with torch.no_grad():
        for (x224, y), (x64, _) in zip(test_loader_224, test_loader_64):
            x224 = x224.to(DEVICE)
            x64 = x64.to(DEVICE)

            p1 = F.softmax(model_lvl2(x224), dim=1)
            p2 = F.softmax(model_lvl3(x64), dim=1)
            p3 = F.softmax(model_lvl4(x224), dim=1)

            avg_prob = (p1 + p2 + p3) / 3
            preds = avg_prob.argmax(1)

            y_true.extend(y.numpy())
            y_pred.extend(preds.cpu().numpy())

    return np.array(y_true), np.array(y_pred)

```

LEVEL-4 ENSEMBLE EVALUATION
